```python
import tensorflow as tf
```

```python
height = [170,180,175,160]
shoes = [260,270,265,255]
# 신발 사이즈와 키사이에 어떤 관계가 있을까 
# y= ax
# ??? = a 170 +b 
# 키를 x  축 신발을 y축으로 해서 그래프상에 점을 딱딲 만들고 직선을 구한다! 
# y=ax+b    신발 = a * 키 + b   /// a와 b 값만 구하면 신발 추측 가능 
```

```python
height =170
shoes = 260

# 예측식 
#shoes = a * hight +b 

# 딥러닝 학습시키기 위해서 a와 b의 정의가 필요 

# w값 가중치 설정 
# 초깃값 아무거나 설정 a,b 좋은 거 나올때 까지 
a= tf.Variable(0.1)
b= tf.Variable(0.2)


#경사하강법을 사용해서 w 변수들을 업데이트 하는 도구 
# 옵티마이저 중 아담 한번 써보자 여러가지 많음 가장 전반적으로 성능이  좋음 
# learning_rate 안 넣어도 되지만 넣음 
opt = tf.keras.optimizers.Adam(learning_rate=0.1)

# 손실함수 만들기
def loss():
  # 실제값 - 예측값 -> 오차를 리턴하면 됨 
  # 근데 -값인 경우 틀려도 값이 좋아지니까 ... 제곱해서 평균내는 식 
  # 손실함수는 실제값과 예측값의 차이를 제곱해서 리턴함 
  shoes= height * a + b
  return tf.square(260 - shoes)


# 경사하강법
# 인수로 손실함수 와 var_list[a,b] <- 업데이트할 가중치 목록 
opt.minimize(loss,var_list=[a,b]) 
# 경사하강을 1번 해줌 = a,b를 1번 수정해줌 
# 0.1 0.2가 아니라 수정된 값을 a,b에 넣어줌 
# 이걸 계속 반복하면됨 오차가 적을 때 까지 

for i in range(300):
    opt.minimize(loss,var_list=[a,b]) 
    print(a.numpy(),b.numpy())


print(height*a+b) # 신발사이즈 260 나옴 홀리
# 그냥 이건 원리시범임 
```
