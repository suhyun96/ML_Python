```python
# 파이썬으로 엑셀 데이터 다루고 싶을 때 사용 
import pandas as pd
data = pd.read_csv('/content/gdrive/MyDrive/code/univ/gpascore.csv')
#print(data)

# 전처리 -> 데이터에 빵꾸난 부분 처리하는 것 간혹 없거나 이상한거 있을 수 있음 

# 빈칸 총 갯수 확인해주기 
data.isnull().sum()
# nan/ 빈칸 제거 해줌 
data=data.dropna()
#data.isnull().sum()

# 빈칸에 임의의 값 채워넣기
#data = data.fillna(100)

# data['gpa'] -> 새로로 gap 열 다 출력 됨 
# data['gpa'].min() -> 해당 열에 최솟값 파악가능 

y_data = data['admit'].values   # values를 통해서 값을 리스트로 반환 
#print(y_data)  # -> 01010101 리스트로 만들어줌 

# x 데이터 
x_data=[]
#data,iterrows() -> 데이터 프레임의 데이터를 가로 한줄씩 출력 가능 즉 한행 ! 
#반복문에서 변수 2개씩 i=인덱스 / rows=행데이터 
for i , rows in data.iterrows():
  x_data.append([ rows['gre'], rows['gpa'], rows['rank'] ] )   # 해당하는 속성들이 리스트에 들어감

#print(x_data)
```

```python
import numpy as np
#import keras
import tensorflow as tf 

# 1. 딥러닝 model 디자인 하기 
# Sequential 쓰면 신경망 레이어들을 쉽게 만들어준다 
model = tf.keras.models.Sequential([ 
                                    #Dense() 에 넣을 파라미터 
                                    # 노드 갯수 / 활성화함수 
                                    # 활성화함수 ex_ sigmoid , tanh, relu , softmax, leakyRelu 등등 
                            tf.keras.layers.Dense(64,activation='tanh'),   # 레이어 만드는 문법 
                                                      # 노드 여러개 있던 hidden layer임 각각 하나의 히든레이어
                            tf.keras.layers.Dense(128,activation='tanh'),
                            tf.keras.layers.Dense(256,activation='tanh'),
                            tf.keras.layers.Dense(1,activation='sigmoid') # 마지막 레이어는 예측결과를 0-1사이로 출력해줘야함 
                            # sigmoid는 0-1 사이로 값을 찌뿌려뜰여서 만들어줌 0.9면 90퍼 확률 
                            ]) #<- 딥러닝 모델 

                            # 64 ,128,1 이거 노드의 갯수임 
                            # 히든레이어 안에 들어갈 노드의 갯수 
                            # 그냥 이거 레이어 갯수. 노드갯수는 맘대로 잘나올 때 까지 함
                            # 보통 2의 제곱으로 하긴함 ... 
                            # 무조건 마지막에는 하나의 노드만 위치하도록 결과물이니까! 0,1 중 하나
                            

# 2. model compile 하기
# compile()함수의 인수는 optimizer , loss , metrics 
# optimizer => 아담을 쓰는군 
# loss 함수는 다 다른데 확률일 경우 binary_crossentropy 결과가 0과 1사이의 분류 /확률에서 사용
model.compile(optimizer='adam' , loss='binary_crossentropy' , metrics=['accuracy'])

# 3. model 함습(fit) 시키기 
# x는 트레이닝 데이터 y는 타겟값 (실제정답) , epochs=10 -> 몇번 학습시킬지 
# fit (x 데이터 , y데이터, epochs=10 )
# 리스트를 넘파이형태로 바꿔야함 
model.fit( np.array(x_data) ,np.array(y_data) ,epochs=1000)

# 예측 
# fit이랑 같은 값인데 테스트해볼 데이터들
predict = model.predict( [ [750, 3.70, 3], [200, 2.20 ,1] ] )
print(predict)
```
